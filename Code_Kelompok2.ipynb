```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Analisis Korelasi Skill yang Diperoleh dari Pekerjaan di Lab SLC terhadap Prestasi Akademik Mahasiswa BINUS Kemanggisan (Responses) - Form Responses 1.csv")
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
pie_charts = ["Jenis Kelamin", "Tahun Kelulusan Kuliah", "Jabatan Anda di Lab SLC sekarang?", "Sudah berapa lama Anda bekerja di Lab SLC?", "Berapa lama rata-rata kamu bekerja di lab SLC setiap harinya?"]
```

```python
import matplotlib.pyplot as plt

# Misal df adalah dataframe kamu dan 'charts' adalah list kolom yang ingin kamu buat pie chart-nya
# Contoh: charts = ['Gender', 'Major']

for col in pie_charts:
    plt.figure(figsize=(6, 6))
    df[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f'Pie Chart of {col}')
    plt.ylabel('')  # Menghilangkan label y
    plt.tight_layout()
    plt.show()
```
